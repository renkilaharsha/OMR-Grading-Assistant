```python
## open txt file
f = open("ground_truth.txt",'r')
list_all_ans = []

dict_options = {'A':0,'B':1,'C':2,'D':3,'E':4}
for i in range(85):
    content = f.readline()
    list_qs = content.strip().split(" ")
    list_all_ans.append(list_qs)
```

```python
list_all_ans
```

```python
list_encoded = []
for i in list_all_ans:
    list_chars= [char for char in i[1]]
    print(list_chars)
    list_encoded.append([dict_options[char] for char in list_chars])
```

```python
list_encoded
```

```python
import numpy as np
from PIL import Image
np_arr_mcq = np.ones((85,5))
print(np_arr_mcq.shape)
```

```python
for idx1, chars in enumerate(list_encoded):
    for char in chars:
         np_arr_mcq[idx1,char] = 200
```

```python
np_arr_mcq
```

```python
img_encode = Image.fromarray(np_arr_mcq)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (3,8),dpi = 100)
plt.imshow(img_encode)
```

```python
from PIL import ImageOps
blank_form_img = Image.open("blank_form.jpg")
gray_image = ImageOps.grayscale(blank_form_img)
gray_img_arr = np.array(gray_image)
```

```python
gray_img_arr[2000:2085,1600:1605] = img_encode
```

```python
plt.figure(figsize = (18,18),dpi = 100)
plt.imshow(gray_img_arr,cmap = 'gray')
```

```python
gray_img_pil = Image.fromarray(gray_img_arr)
gray_img_pil.save("barcoded_img.jpg")
```

```python
barcoded_img = Image.open("barcoded_img.jpg")
```

```python
barcoded_image = ImageOps.grayscale(barcoded_img)
barcoded_img_arr = np.array(barcoded_image)
```

```python
barcoded_img_arr[2000:2085,1600:1605][barcoded_img_arr[2000:2085,1600:1605] > 100] = 200
barcoded_img_arr[2000:2085,1600:1605][barcoded_img_arr[2000:2085,1600:1605] < 100] = 1
```

```python
barcoded_img_arr[2000:2085,1600:1605]
```

```python
comparison = np_arr_mcq == barcoded_img_arr[2000:2085,1600:1605]
equal_arrays = comparison.all()
print(equal_arrays)
```

```python
barcode = barcoded_img_arr[2000:2085,1600:1605]
dict_reverse = {0:'A',1:'B',2:'C',3:'D',4:'E'}
list_final_ans = []
for i in range(85):
    str_init = ''
    for j in range(5):
        if barcode[i,j] == 200:
            str_init+=dict_reverse[j]
    list_final_ans.append(str_init)
```

```python
list_final_ans
```

```python
extract_text = open("extract.txt",'w')
for idx,ans in enumerate(list_final_ans):
    extract_text.write(f"{idx+1} {ans}\n")
extract_text.close()
```
